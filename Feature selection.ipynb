```python
# CONVERTING DIAGNOSIS COLUMN TO 1 & 0

from sklearn.preprocessing import  LabelEncoder
le=LabelEncoder()
breast_cancer_df['diagnosis']=le.fit_transform(breast_cancer_df['diagnosis'])
```

```python
# SELECTING X AND Y FEATURES

X=breast_cancer_df.drop('diagnosis',axis=1)
Y=breast_cancer_df['diagnosis']
fea_names=X.columns
fea_names.value_counts().count()
```

```python

from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier
fea_sel=RandomForestClassifier()

rfe=RFE(fea_sel,n_features_to_select=10)
fit=rfe.fit(X,Y)

print(f'THE SELECTED FEATURES ARE : {fit.n_features_}')

print(f'the ranking of featurez are :{fit.ranking_}')

print(f'the {fit.support_}')

selected_feature_names = [fea_names[i] for i in range(len(fea_names)) if fit.support_[i]]

print(selected_feature_names)
DF1=pd.DataFrame({'SELETEC FEATURES':selected_feature_names})

DF1
SELECT_DF=breast_cancer_df[['texture_mean', 'perimeter_mean', 'area_mean', 'concavity_mean', 'concave points_mean', 'radius_worst', 'texture_worst', 'perimeter_worst', 'area_worst', 'concave points_worst','diagnosis']]
SELECT_DF
```

```python
# SPLITING Data into TRAING AND TESTING SET
from sklearn.model_selection import train_test_split
X_TRAIN,X_TEST,Y_TRAIN,Y_TEST=train_test_split(X,Y,test_size=0.2,random_state=20)
```

```python
# PERFORMING STANDIZATION FOR SCALING THE VALUES

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_TRAIN=sc.fit_transform(X_TRAIN)
X_TEST=sc.transform(X_TRAIN)
```

```python
# BUDLDING THE RandomForestClassifier MODEL

from sklearn.ensemble import RandomForestClassifier
MODEL_RF=RandomForestClassifier(n_estimators=10,max_depth=4,min_samples_split=5,min_samples_leaf=4, random_state=10)
MODEL_RF.fit(X_TRAIN,Y_TRAIN)
```

```python
# EVALUATING THE MODEL feature_importance 
feature_importance=MODEL_RF.feature_importances_
feature_name=X.columns
```

```python
# CREATING DATAFRAME FOR feature_importance

feature_df=pd.DataFrame({'FEATURE':feature_importance,
                         'NAMES':feature_name}).sort_values(by='FEATURE',ascending=False).sort_values(by='FEATURE',ascending=False)
```

```python
# PLOTING GRAPH FOR VISULIAZTION FOR feature_importance

plt.figure(figsize=(20,6))

sns.barplot(data=feature_df,y=feature_df['NAMES'],x=feature_df['FEATURE'])
```
